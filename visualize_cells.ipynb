```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
#Having a hint of the data
metadata = pd.read_table("data/HCC1806_SmartS_MetaData.tsv")
print(metadata.columns)
metadata.head()
```

```python
#Visualise the share of cells cultivated in Normal/Hypoxia conditions
mpl.hist(metadata["Condition"])
```

```python
#Checking that a single cell is cultivated in a single environment
print("Length of unique Cell names: ", len(set(metadata["Cell name"])) )
print("Length of unique Cell ids : ", len(set(metadata["Filename"])))
```

```python
#Changing to categorical variable
# "Norm" --> "Normal" --> 1
# "Hypo" --> "Hypoxia"--> 0
metadata["Condition"] = metadata["Condition"].apply(lambda x: 1 if x == "Normo" else x)
metadata["Condition"] = metadata["Condition"].apply(lambda x: 0 if x == "Hypo" else x)
metadata
```
