```python
import numpy as np
import pandas as pd

## MODIFY NAMES OF COLUMNS AND CREATE TARGET COLUMN

import matplotlib.pyplot as plt
data = pd.read_csv("data/HCC1806_SmartS_Unfiltered_Data.txt", sep= " ")
original_data = data.copy()
col = list(data.columns)

params = []
for i,k in enumerate(col):
    col[i] = k.split('.')
    col[i] = col[i][2]
    params.append(col[i].split('_'))
    params[-1] = params[-1][1]

data.columns = col

# data = data.T
# data['target'] = params
    
data
```

```python
data = data.T
cells =['PCRPlate1A10_Normoxia_S123_Aligned', 'PCRPlate1A12_Normoxia_S26_Aligned', 'PCRPlate1A1_Hypoxia_S97_Aligned', 'PCRPlate1A2_Hypoxia_S104_Aligned']
data
```

```python
## CELLS THAT EXPRESSESS FEW GENES

data_num = data.select_dtypes(include=[np.number])
cells_few_genes = []
for i in data_num.columns:
    if data_num[[f'{i}']].sum()[0] < 100000:
        cells_few_genes.append(i)



print(cells_few_genes, '\n',len(cells_few_genes))
plt.figure(figsize=(24,15))
plt.ticklabel_format(style='plain')
plt.bar(data.columns, data_num.sum(axis=0))
plt.xlabel('Cells', fontsize= 30)
plt.ylabel('Number of genes expressed', fontsize= 30)
plt.grid(visible=True, axis= 'y')
plt.show()
```

```python
## GENES EXPRESSED BY SOME CELLS
data[cells].hist(bins=150, figsize=(18, 10))
plt.show()
```
